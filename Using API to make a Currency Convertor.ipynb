## Incorporating parameters in the GET request

### to specify a parameter, we incorporate it in the URL itself:
####    http://......?par1=Value1&par2=Value2&par3=Value3 
### In the following case a '&' was used instead of a '?' to begin the parameter specification

```python
param_url=baseUrl+"&symbols=USD,GBP"
param_url
```

```python
new_response=r.get(param_url)
```

```python
data=new_response.json()
```

```python
data
```

```python
data['base']
```

```python
data['date']
```

## Calculating conversion rates using the APIs paramters, "base" and "symbols"

```python
param_url=baseUrl+"?symbols=GBP&base=USD"
param_url
```

```python
data=r.get(param_url).json()
```

```python
data
```

```python
usd_gbp=data['rates']['GBP']
usd_gbp
```

## Building a simple currency converter 

```python
def currency_exchange(base,target,amount):
    url="https://api.exchangerate.host/convert"+"?from="+base.upper()+"&to="+target.upper()+"&amount="+amount
    dut=r.get(url).json()
    return dut['result']
```

```python
a=input('Enter the base currency:\n ')
b=input('Enter the currency you would like to convert to:\n')
c=input('Enter the amount you would like to convert to:\n')
print(' The converted value is',currency_exchange(a,b,c))
```
