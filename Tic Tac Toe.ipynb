```python
#tic tac
import random
def db(board):
    print('   |   |')
    print(' ' + board[7] + ' | ' + board[8] + ' | ' + board[9])
    print('   |   |')
    print('-----------')
    print('   |   |')
    print(' ' + board[4] + ' | ' + board[5] + ' | ' + board[6])
    print('   |   |')
    print('-----------')
    print('   |   |')
    print(' ' + board[1] + ' | ' + board[2] + ' | ' + board[3])
    print('   |   |')
def inputplayerLetter():
    letter=''
    while not (letter=='X') or (letter=='O'):
        print("Do you want to be 'X' or 'O'?")
        letter=input().upper()
    if letter=='X':
        return ['X','O']
    else:
        return ['O','X']
def firstchance():
    a=random.randint(0,1)
    if a ==0:
        return 'computer'
    else:
        return 'player'
def playagain():
    print("Do you want to play again?, Enter Y or N:")
    b=input().upper()
    return b.startswith('Y')
def movingmove(board,letter,move):
    board[move]=letter
def copyboard(board):
    dupe=[]
    for i in board:
        dupe.append(i)
    return dupe
def wins(bo,le):
    return ((bo[7] == le and bo[8] == le and bo[9] == le) or # across the top
            (bo[4] == le and bo[5] == le and bo[6] == le) or # across the middle
            (bo[1] == le and bo[2] == le and bo[3] == le) or # across the bottom
            (bo[7] == le and bo[4] == le and bo[1] == le) or # down the left side
            (bo[8] == le and bo[5] == le and bo[2] == le) or # down the middle
            (bo[9] == le and bo[6] == le and bo[3] == le) or # down the right side
            (bo[7] == le and bo[5] == le and bo[3] == le) or # diagonal
            (bo[9] == le and bo[5] == le and bo[1] == le)) # diagonal
def isfreespace(board,move):
    return board[move]==' '
def getPlayermove(board):
    move=' '
    while move not in '1 2 3 4 5 6 7 8 9'.split() or not isfreespace(board,int(move)):
        print("What is your next move?")
        move=input()
    return int(move)
def chooserandomove(board,movesl):
    possible=[]
    for i in movesl:
        if isfreespace(board,i):
            possible.append(i)
        if len(possible)!=0:
            return random.choice(possible)
        else:
            return None
#Game Engine
def getcompmove(board,comp):
    if comp=='X':
        play='O'
    else:
        play='X'
    for i in range(1,10):
        copy=copyboard(board)
        if isfreespace(copy,i):
            movingmove(copy,comp,i)
            if wins(copy,comp):
                return i
    for i in range(1,10):
        copy=copyboard(board)
        if isfreespace(copy,i):
            movingmove(copy,play,i)
            if wins(copy,play):
                return i
    move=chooserandomove(board,[1,3,7,9])
    if move!=None:
        return move
    if isfreespace(board,5):
        return 5
    move=chooserandomove(board,[2,4,6,8])
def isfull(board):
    for i in range(1,10):
        if isfreespace(board,i):
            return False
    return True

print("Welcome to tic tit tot")
while True:
    board=[' ']*10
    play,comp=inputplayerLetter()
    turn=firstchance()
    print(turn+' has the first turn')
    gplay=True
    while gplay:
        if turn =='player':
            db(board)
            move=getPlayermove(board)
            movingmove(board,play,move)
            if wins(board,play):
                db(board)
                print("You wins, Computer loses")
                gplay=False
            else:
                if isfull(board):
                    db(board)
                    print('Game tied')
                    break
                else:
                    turn='computer'
        else:
            move=getcompmove(board,comp)
            movingmove(board,comp,move)
            if wins(board,comp):
                db(board)
                print("Computer wins, you lose")
                gplay=False
            else:
                if isfull(board):
                    db(board)
                    print("Game tied")
                    break
                else:
                    turn='player'
    if not playagain():
        break
        

                



        
    
    
        
    
    
    
        
```
